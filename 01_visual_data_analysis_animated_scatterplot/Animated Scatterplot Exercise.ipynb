```python
# Step 11 To create an animated scatterplot showing the correlation between life expectancy 
# and fertility, create one plot for each year between 1960 and 2015 
# Write a for loop that goes through each year and produces a scatterplot. 
# Step 12 Add another dimension to the visualization by utilizing the population data.
# Adjust the plot, so that the size of the symbols represent the population from the gapminder dataset.

import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')

for current_year in range(1960, 2015):
    df_subset = df.loc[df['year'] == current_year]
    ax = sns.scatterplot(x='life-expectancy', y='fertility_rate', hue='continent', size= 'population', sizes=(10,1000), data=df_subset, alpha=0.8) 
    h,l = ax.get_legend_handles_labels()
    plt.legend(h[1:7],l[1:7],loc='center left',bbox_to_anchor=(1,0.7), borderaxespad=2., fontsize=8, title= "Continents:");  
    plt.axis((0, 100, 0, 10))
    plt.title(f"Population of countries in {current_year}")
    plt.xlabel('Life expectancy')
    plt.ylabel('Fertility Rate')
    plt.savefig(f'lifeexp_{current_year}.png', dpi=150, bbox_inches='tight')
    # plt.show()
    plt.close()
```
